```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
img = cv.cvtColor(cv.imread('sc.png'), cv.COLOR_BGR2RGB)
rows, cols, channels = img.shape
plt.imshow(img)
print(rows, cols, channels)
```

Y = 0.2126 R + 0.7152 G + 0.0722 B


```python
hue_img = img[:,:,0]*0.2126 + img[:,:,1]*0.7152 + img[:,:,2]*0.0722
```

```python
plt.imshow(hue_img, cmap='gray')
```

```python
sorted_img = np.zeros_like(img)
for i in range(cols):
    col_rgb = img[:, i, :]
    col_hue = col_rgb[:,0]*0.2126 + col_rgb[:,1]*0.7152+col_rgb[:,2]*0.0722
    hue_sort_inds = col_hue.argsort()
    col_rgb_sorted = col_rgb[hue_sort_inds[::-1]]
    sorted_img[:,i] = col_rgb_sorted
```

```python
plt.imshow(sorted_img)
cv.imwrite('sc_out.png', cv.cvtColor(sorted_img,cv.COLOR_BGR2RGB))
```
